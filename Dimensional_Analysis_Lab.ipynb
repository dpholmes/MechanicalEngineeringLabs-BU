## Background
As you might recall from physics, a pendulum is often modeled as a [harmonic oscillator](https://en.wikipedia.org/wiki/Harmonic_oscillator), such that the angle $\theta$ of the pendulum oscillates sinusoidally with time $t$.  In our Jupyter notebook, we can use [LaTeX](https://en.wikibooks.org/wiki/LaTeX/Mathematics) (check out the source code) to write an equation of the angle: 
\begin{equation}
\theta = \theta_0\cos \left(2\pi\frac{t}{T}\right), 
\end{equation}
which will be implemented in the Python code below as:
````
theta = theta_0*np.cos(2*np.pi*t/T)
````
Here $\theta_0$ is the initial angle from which the pendulum is released and $T$ is the amount of time it takes for the pendulum to return to its starting point, often refered to as the period.

Below is some Python code that illustrates how the angle changes with time. You are encouraged to experiment with the parameters in the code to see how the plots respond. This code also serves as a Python template when plotting multiple functions with a title, legend, and axis labels.

`````python
# First, we import some Python libraries. This only needs to be done once in the notebook.
import numpy as np                        # Working with arrays/matrices
import matplotlib.pyplot as plt           # Plotting similar to MATLAB
`````

`````python
# It doesn't really matter here if we use degrees or radians, but it might matter later. 
# Remember that radians are dimensionless.
theta_0 = 15
Ts = [0.5, 2, 4]
t = np.linspace(0,4,200)

for T in Ts:
  theta = theta_0*np.cos(2*np.pi*t/T)
  plt.plot(t,theta,label='T = {} s'.format(T))

plt.legend()
plt.xlabel("Time (s)")
plt.ylabel("Angle (deg)") #Change the units if you decide to use radians
plt.title("Simple pendulum motion for different periods")
plt.show()
`````
